```python
def add(a,b):
    return a+b
def sub(a,b):
    return a-b
def mul(a,b):
    return a*b
def div(a,b):
    return a/b
a = int(input("Enter a = "))
b = int(input("Enter b = "))
o = input("Enter + or - or / or * ")
if o == '+':
    print(add(a,b))
elif o == '-':
    print(sub(a,b))
elif o =='*':
    print(mul(a,b))
elif o == '/':
    print(div(a,b))
else:
    print('Invalid operator')

    
```

```python
# recursive function
def hello():
    print("Hello World")
    hello()
hello()
```

```python
# recursive function (For stop)
def hello():
    print("Hello World")
    x = input("Enter y for more print ")
    if x == 'y':
        hello()
    
hello()
```
